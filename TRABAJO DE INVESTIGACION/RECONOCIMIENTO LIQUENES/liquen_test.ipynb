```python
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
model = YOLO("best_liquen_v2.pt")
```

```python
results = model("./image_test/", conf=0.25)
for res in results :
    res.show()
    # Obtener la máscara (primer objeto segmentado)
    mask = res.masks.data[0].cpu().numpy()

    # # Binarizar (por seguridad numérica)
    mask_bin = (mask > 0.5).astype(np.uint8)

    # Conteo de píxeles
    area_pixels = np.count_nonzero(mask_bin)
    area_escalada = area_pixels*0.25
    print(f"Área en píxeles (conteo directo): {area_pixels}")
    print(f"Área escalada: ", area_escalada)
```
