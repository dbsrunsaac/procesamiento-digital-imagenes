```python
import cv2
import numpy as np
from PanelControl import *
```

```python
def empty(a) :
    pass

path = 'colors.png'

cv2.namedWindow("TrackBars")
cv2.resizeWindow("TrackBars", 640, 480)
cv2.createTrackbar("Hue Min","TrackBars", 0, 179, empty)
cv2.createTrackbar("Hue Max","TrackBars", 0, 179, empty)
cv2.createTrackbar("Sat Min","TrackBars", 0, 255, empty)
cv2.createTrackbar("Sat Max","TrackBars", 0, 255, empty)
cv2.createTrackbar("Val Min","TrackBars", 0, 255, empty)
cv2.createTrackbar("Val Max","TrackBars", 0, 255, empty)

cv2.createTrackbar("Val Ker", "TrackBars", 0, 15, empty)
cv2.createTrackbar("Val Ero", "TrackBars", 0, 15, empty)
cv2.createTrackbar("Val Dil", "TrackBars", 0, 15, empty)


while True :
    img = cv2.imread(path)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    h_min = cv2.getTrackbarPos("Hue Min", "TrackBars")
    h_max = cv2.getTrackbarPos("Hue Max", "TrackBars") 
    s_min = cv2.getTrackbarPos("Sat Max", "TrackBars") 
    s_max = cv2.getTrackbarPos("Sat Min", "TrackBars") 
    v_min = cv2.getTrackbarPos("Val Max", "TrackBars") 
    v_max = cv2.getTrackbarPos("Val Min", "TrackBars")
    v_ker = cv2.getTrackbarPos("Val Ker", "TrackBars")
    v_ero = cv2.getTrackbarPos("Val Ero", "TrackBars")
    v_dil = cv2.getTrackbarPos("Val Dil", "TrackBars")

    lower = np.array([h_min, s_min, v_min])
    upper = np.array([h_max, s_max, v_max])
    mask = cv2.inRange(img_hsv, lower, upper)

    kernel = np.ones((v_ker, v_ker), np.uint8)

    img_erode = cv2.erode(mask, kernel, iterations=v_ero)
    img_dilatacion = cv2.dilate(img_erode, kernel, iterations=v_dil)

    img_result = cv2.bitwise_and(img, img, mask=img_dilatacion)

    # Archivo de configuración
    config = {
        'imagenes' : [img, img_hsv, mask, img_erode, img_dilatacion, img_result],
        'etiquetas' : ["Original", "HSV", "Mask", "Erocion", "Dilatacion", "Img Resultante"],
        'filas': 2,
        'columnas': 3,
        'tamaño_base' : 250,
        'escala_panel' : 100
    }
    
    # Configuración

    panel = crear_panel(**config)
    cv2.imshow('Panel', panel)

    # cv2.imshow("Img ori", img)
    # cv2.imshow("Img HSV", img_hsv)
    # cv2.imshow("Img Mask", mask)
    # cv2.imshow("Img Erode", img_erode)
    # cv2.imshow("Img Dilate", img_dilatacion)
    # cv2.imshow("Img Result", img_result)

    if cv2.waitKey(1) & 0xFF == 27 :
        break

cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
ancho = cap.get(cv2.CAP_PROP_FRAME_WIDTH) 
alto = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

fps_cam = cap.get(cv2.CAP_PROP_FPS)

print(f"FPS: {fps_cam}, ancho: {ancho}, alto: {alto}")
while True :


    _, frame = cap.read()

    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == 27 :
        break

cap.release()
cv2.destroyAllWindows()
```
