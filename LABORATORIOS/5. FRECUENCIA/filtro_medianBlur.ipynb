```python
import cv2
import numpy as np
```

#### medianBlur
Es uno de los mejores filtros para eliminar ruido

```python
image = cv2.imread('herby.jfif')

# Aplicación del filtro gaussiano (tamaño del kernel 5x5)
gaussian_filtered = cv2.medianBlur(image, 5)

cv2.imshow('Imagen Original', image)
cv2.imshow('Filtro median blur (5x5)', gaussian_filtered)

cv2.imwrite('herby.jpg', gaussian_filtered)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
