```python
import cv2
import numpy as np
import plotly.graph_objects as go
```

```python
img = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)

dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]) + 1)

selected_colorscale = 'Jet'

fig = go.Figure(data=[go.Surface(z=magnitude_spectrum, colorscale=selected_colorscale)])

fig.update_layout(
    title='Espectro de Magnitud en 3D',
    scene=dict(
        xaxis_title='Frecuencia Horizontal (u)',
        yaxis_title='Frecuencia Vertical (v)',
        zaxis_title='Magnitud (log)',
        aspectmode='cube'
    ),
    autosize=False,
    width=800,
    height=700,
    margin=dict(l=65, r=50, t=90)
)
fig.show()
```
