```python
import cv2
import numpy as np
```

```python
image = cv2.imread('herby.jpg')

laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=7)

laplacian = np.absolute(laplacian)
laplacian = np.uint8(laplacian/np.max(laplacian)*255)

cv2.imshow('Imagen Original', image)
cv2.imshow('Bordes laplaciano', laplacian)

cv2.imwrite('laplaciano.jpg', laplacian)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
