```python
import cv2
import numpy
from PanelControl import *
```

```python
url = 'https://10.10.37.61:8000/video'
vid = cv2.VideoCapture(url)

while True : 
    _, frame = vid.read()
    frame_linear = cv2.blur(frame, (7, 7))
    frame_gaussian_blue = cv2.GaussianBlur(frame, (7,7), 20)
    frame_median = cv2.medianBlur(frame, 15)
    
    # * Sobel
    sobelx = cv2.Sobel(frame, cv2.CV_64F, 1, 0, ksize=3)
    sobelx = np.absolute(sobelx)
    # Vuelve a realizar la conversión a 8bits
    sobelx = np.uint8(sobelx / np.max(sobelx)*255)

    # Sobel en dirección y (detecta bordes horizontales)
    sobely = cv2.Sobel(frame, cv2.CV_64F, 0, 1, ksize=3)
    sobely = np.absolute(sobely)
    sobely = np.uint8(sobely/np.max(sobely)*255)

    # Sobel combinado (aproximando la magnitud total del gradiente)
    frame_sobel_combined = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

    # Laplacian
    laplacian = cv2.Laplacian(frame, cv2.CV_64F, ksize=7)
    laplacian = np.absolute(laplacian)
    frame_laplacian = np.uint8(laplacian/np.max(laplacian)*255)
    


    # Archivo de configuración 
    config = {
        'imagenes' : [frame, frame_linear, frame_gaussian_blue, frame_median, frame_sobel_combined, frame_laplacian],
        'etiquetas' : ["Original", "Linear filter", "Gaussian", "Median blur", "Sobel", "Laplacian"],
        'filas': 2,
        'columnas': 4,
        'tamaño_base' : 250,
        'escala_panel' : 100
    }

    # Configuración

    panel = crear_panel(**config)
    cv2.imshow('Panel Filtros', panel)

    if cv2.waitKey(30) & 0xFF == 27 :
        break


vid.release()
cv2.destroyAllWindows()

    
```
