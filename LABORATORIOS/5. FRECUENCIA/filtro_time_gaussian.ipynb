```python
import cv2
import numpy as np
```

```python
image = cv2.imread('herby.jfif')

# Aplicación del filtro gaussiano (tamaño del kernel 5x5)
gaussian_filtered = cv2.GaussianBlur(image, (7,7), 20)

cv2.imshow('Imagen Original', image)
cv2.imshow('Filtro Gaussiano (5x5)', gaussian_filtered)

cv2.imwrite('blurg.jpg', gaussian_filtered)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
