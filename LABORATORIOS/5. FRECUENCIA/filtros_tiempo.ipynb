```python
import cv2
import numpy as np
```

```python
image = cv2.imread('herby.jfif')

# Aplicación del filtro de primedio (tamaño de kernel) 7x7
mean_filtered = cv2.blur(image, (5, 5))

cv2.imshow('Imagen Original', image)
cv2.imshow('Filtro de Promedio (5x5)', mean_filtered)

cv2.imwrite('blurg.jpg', mean_filtered)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
