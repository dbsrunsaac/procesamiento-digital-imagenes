```python
import cv2
import numpy as np
```

```python
image = cv2.imread('herby.jpg')

# Aplicación del operador Sobel
# ddepth: profundidad de la imagen de salida (cv2.CV_64F para flotante para manejar valores negativos)
# dx, dy: Orden de la derivada (1 para x, 0 para Y o viceversa)
# ksize: Tamaño del kernel de Sobel(E.g, 3 para 3x3)

# Sobel para dirección
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobelx = np.absolute(sobelx)

# Vuelve a realizar la conversión a 8bits
sobelx = np.uint8(sobelx / np.max(sobelx)*255)

# Sobel en dirección y (detecta bordes horizontales)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobely = np.absolute(sobely)
sobely = np.uint8(sobely/np.max(sobely)*255)

# Sobel combinado (aproximando la magnitud total del gradiente)
# Los pesos deben sumar 1, por lo tanto uno es el complemento a 1 del otro
sobel_combined = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
cv2.imshow('Imagen Original', image)
# cv2.imshow('Sobel X (Bordes Verticales)', sobelx)
# cv2.imshow('Sobel Y (Bordes Horizontales)', sobely)
cv2.imshow('Sobel combinado', sobel_combined)

cv2.imwrite('Sobel.jpg', sobel_combined)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
