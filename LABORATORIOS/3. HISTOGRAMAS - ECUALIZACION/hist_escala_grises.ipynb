```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Cargar las imagenes en escalas de grises
imagen_gris = cv2.imread("./simba.jpg", cv2.IMREAD_GRAYSCALE)

# Calculo del histograma
histograma = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])

# Crear la figura con 2 sbuplots
plt.figure(figsize=(12, 5))

# Subplot 1: Mostrar imagen
plt.subplot(1, 2, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title("Imagen a escala de grises")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.plot(histograma, color='red')
plt.title('Histograma')
plt.xlabel('Niveles de intensidad y/o R.D.')
plt.ylabel('Pixeles')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
