```python
import cv2
import matplotlib.pyplot as plt
```

```python
"""
    Recordar que matlplotlib trabaja con RGB y cv2 con BGR
    (Por ello se realiza el cambio)
"""

imagen_color = cv2.imread("./simba.jpg")
imagen_rgb = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB)

# color red
r, g, b = cv2.split(imagen_rgb)

hist_red = cv2.calcHist([r], [0], None, [256], [0, 256])
hist_green = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_blue = cv2.calcHist([b], [0], None, [256], [0, 256])

# hist green

# hist blue
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title("Imagen RGB")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.plot(hist_red, color='red', label='rojo')
plt.plot(hist_green, color='green', label='green')
plt.plot(hist_blue, color='blue', label='blue')
plt.title("Histograma")
plt.xlabel("Niveles de intensidad y/o R.D.")
plt.ylabel("Frecuencia")
plt.xlim([0, 256])
plt.legend()

plt.tight_layout()
plt.show()

```

### Optimizar el cdigo previo
Esto se realiza mediante el uso de un bucle for

```python
img_color = cv2.imread("./simba.jpg")
img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

plt.figure( figsize=(12, 5))
plt.subplot(1, 2, 1)

plt.imshow(img_rgb)
plt.title("Imagen a color")
plt.axis('off')

colores = ["red", "green", "blue"]
etiquetas = ["rojo", "verde", "azul"]
plt.subplot(1, 2, 2)
for i in range(0, 3) :
    capa = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    plt.plot(capa, color=colores[i], label=etiquetas[i])

plt.title("Histograma por color")
plt.xlabel("Nivel de intensidades y/o R.D.")
plt.ylabel("Frecuencia")
plt.xlim([0, 256])
plt.legend()

plt.tight_layout()
plt.show()
```
